```python
"""
Course: CSCI490
Author: Jesus Lua

Function: This program implements data detection on plotting the data using various
techniques, by using the famous iris data set to read and plot the data, as well as
reading link data to build plots and applying transparency and jittering to those plots
to determine between good or bad visualization.
"""

import numpy as np
import matplotlib.pyplot as plt

"""
This first graph graphs two components of data, and we compare the two within the
year time line and how they relate using a line plot graph.
"""
x_years = np.array([1999, 2000, 2001, 2002, 2003,2004, 2005, 2006, 2007, 2008, 2009])
y_barrels = np.array([96,110,103,127,60,54,43,36,20,11,22])
z_deaths = np.array([76, 74, 76, 87,66,59, 63, 60, 55, 52,46])

fig, ax1 = plt.subplots(figsize=(12,4))


color = 'tab:red'
ax1.set_title('US crude oil imports from Norway\n correlates with\n ', color='red')
ax1.set_ylabel('US crude oil imports from Norway', color=color)
line1, = ax1.plot(x_years, z_deaths, color=color)
ax1.tick_params(axis='y', labelcolor=color)

#The twinx function allows us to make a copy of ax1 and be able to plot on both
#y axis inorder to compare the stats.
ax2 = ax1.twinx()

ax2.set_title('Drivers killed in collision with railway train', color='black')
ax2.set_ylabel('Railway train collisions', color='black')
line2, = ax2.plot(x_years, y_barrels, color='black')
ax2.tick_params(axis='y', labelcolor='black')


fig.tight_layout()
plt.xlim(1999, 2009)

ax2.legend((line1, line2), ("US crude imports from Norway", "Railway train collisions"),
          loc="lower center")

plt.show()
```

```python
import urllib
import os
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd


"""
In this part of the program, we read a file link and decode the link to plot the data.
Once the data is plotted we apply a few techniques, transparency and jittering inorder
to show from good to bad visualization. From adding the perfect amount of jittering and
too little to too much jittering. Transperancy allowing us to see overplotted points.
We will scatter the data points to better visualization.
"""

#The fuction below reads and decodes the file stored in bytes into a file of strings
url = "https://raw.githubusercontent.com/pratoolbharti/NIU/master/Applied_Machine_Learning_S20/Homework-1/mpg.csv"
file = urllib.request.urlopen(url).read().decode("utf-8")

new = ""
a = 0
b = 0

"""
This function iterates through file by character and stores it as a string
into an array of strings for later use. 
Also resetting the string variable inorder to store the next string.
"""
cols = 250
rows = 250

arr = [[0 for i in range(cols)] for j in range(rows)] 

for x in file:
    if b == 11:
        a += 1
        b = 0
    if x != ',':
        if x != '"': 
            new += x
    if x == ',':
        arr[a][b] = new
        new = ""
        b += 1

#Set of arrays to contain data as strings for FWD,RWD,and 4WD
newarr= []
newarr2 = []
newarr3 = []

it = 1
#Adding the strings and storing it into the arrays
for it in range(234):
    newarr.append(arr[it][3])
    
for it in range(234):
    newarr2.append(arr[it][8])
for it in range(234):
    newarr3.append(arr[it][7])

#list of arrays for every feature
displ = []
fuelcity = []
drv_arr = []

#Creating list of arrays for types of drv
rwd_x = []
fwd_x = []
fourwd_x = []

rwd_y = []
fwd_y = []
fourwd_y = []

#We convert the string data and convert it into float numbers to use for plotting
x = 0
for x in newarr:
    if x != 'displ':
        f = float(x)
        displ.append(f)

for x in newarr2:
    if x != 'cty':
        f = float(x)
        fuelcity.append(f)
for x in newarr3:
    if x != 'drv':
        #f = float(x)
        drv_arr.append(x)

t = 0  
#Splitting the data into x and y points to plot FWD,RWD,and 4WD individually
for x in drv_arr:
    if x == 'f':
        fwd_x.append(displ[t])
        fwd_y.append(fuelcity[t])
    if x == 'r':
        rwd_x.append(displ[t])
        rwd_y.append(fuelcity[t])
    if x == '4':
        fourwd_x.append(displ[t])
        fourwd_y.append(fuelcity[t])
    t += 1

plt.figure()

#ax = plt.subplot(221)
#plt.scatter(fwd_x,fwd_y)
#plt.show()

f = plt.figure(figsize=(10,10))
ax = f.add_subplot(221)
fwd = ax.scatter(fwd_x,fwd_y, color='orange',linewidth=1)
rdw = ax.scatter(rwd_x,rwd_y, color='cyan',linewidth=1)
fourwd = ax.scatter(fourwd_x,fourwd_y, color='black',linewidth=1)
plt.ylabel("fuel economy (mpg)")
plt.xlabel("displacement(I)")
plt.legend((fwd, rwd, fourwd), ('FWD', 'RWD', '4WD'),title="drive train")


ax2 = f.add_subplot(222)
fwd = ax2.scatter(fwd_x,fwd_y, color='orange',alpha = 0.4,linewidth=1)
rdw = ax2.scatter(rwd_x,rwd_y, color='cyan',alpha = 0.4,linewidth=1)
fourwd = ax2.scatter(fourwd_x,fourwd_y, color='black',alpha = 0.4,linewidth=1)
plt.ylabel("fuel economy (mpg)")
plt.xlabel("displacement(I)")
plt.legend((fwd, rwd, fourwd), ('FWD', 'RWD', '4WD'),title="drive train")

ax3 = f.add_subplot(223)
fwd = ax3.scatter(fwd_x,fwd_y, color='orange',alpha = 0.4,linewidth=1)
rwd = ax3.scatter(rwd_x,rwd_y, color='cyan',alpha = 0.4,linewidth=1)
fourwd = ax3.scatter(fourwd_x,fourwd_y, color='black',alpha = 0.4,linewidth=1)
plt.ylabel("fuel economy (mpg)")
plt.xlabel("displacement(I)")
plt.legend((fwd, rwd, fourwd), ('FWD', 'RWD', '4WD'),title="drive train")
jitterdots = ax3.collections[-1]
offsets = jitterdots.get_offsets()
jittered_offsets = offsets + np.random.uniform(0, 1, offsets.shape)
jitterdots.set_offsets(jittered_offsets)

ax4 = f.add_subplot(224)
fwd = ax4.scatter(fwd_x,fwd_y, color='orange',alpha = 0.3,linewidth=1)
rwd = ax4.scatter(rwd_x,rwd_y, color='cyan',alpha = 0.3,linewidth=1)
fourwd = ax4.scatter(fourwd_x,fourwd_y, color='black',alpha = 0.3,linewidth=1)
plt.ylabel("fuel economy (mpg)")
plt.xlabel("displacement(I)")
plt.legend((fwd, rwd, fourwd), ('FWD', 'RWD', '4WD'),title="drive train")
jitterdots = ax4.collections[-3]
offsets = jitterdots.get_offsets()
jittered_offsets = offsets + np.random.uniform(0, 1, offsets.shape)
jitterdots.set_offsets(jittered_offsets)
plt.show()

```
